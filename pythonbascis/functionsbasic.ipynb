```python
#To write a function that returns name with greeting
def greet(name):
    return "Hello,"+name
print(greet("Bhumika"))
```

```python
#To print a greeting without argument
def greet():
    return "Hello World"

print(greet())
```

```python
#To define a function add(a,b) that returns sume of a and b
def add(a,b):
    total= a+b
    return total

print(add(2,3))
```

```python
#To write a function that retuens the absolute value of n
def absolute(n):
    if n>0:
        return n
    else:
        return -(n)

print(absolute(-5))
```

```python
#To write a function that returns n squared
def sq(n):
    square=n*n
    return square

print(sq(2))
```
